```python
import os
import json
import pandas as pd
import shutil 
```

```python
df = pd.read_csv("train/train.csv") 
print(df.describe())
print(df)
df = df[df["HeadCount"]<=5]
df = df[df["HeadCount"]>=2]
df.describe()
```

```python
filenames = os.listdir('./train/image_data')
sample_json1 = []
for i in filenames:
    #print(df["Name"].values)
    if (i in df["Name"].values):
        a = int(df[df["Name"]==i]["HeadCount"].values[0])
        #print("a:",type(a))
        sample_json1.append(
            {
                'image_id' : i,
                'label' : a
            }
        )
    else:
        continue    
sample_json1
```

```python
filenames = os.listdir('./train/image_data')
for i in filenames:
    if (i in df["Name"].values):
        a = int(df[df["Name"]==i]["HeadCount"].values[0])
        #print("a:",type(a))
        src = './train/image_data/'
        dir1 = './train1/'
        shutil.move(src+i,dir1+i)
    else:
        continue
```

```python
with open('data1.json', 'w') as outfile:
    json.dump(sample_json1, outfile, indent=4, sort_keys=True)
```
