```python
# PDF
dist_norm = stats.norm(loc=0, scale=1) #평균 0이고 표준편차 1인 정규분포 객체 만들기

x=np.linspace(-5,5,100)
y=dist_norm.pdf(x)

plt.plot(x,y)
plt.title("normal distribution : PDF")
plt.show()
```

```python
# CDF

y2= dist_norm.cdf(x)
plt.plot(x,y2)
plt.title("normal distribution : CDF")
plt.show()
```

```python
# percent point function

a = dist_norm.ppf(0.05)
b = dist_norm.ppf(0.95)

print("The percent points of a and b are {} and {}".format(a,b))
```

```python
# sampling
sample_norm = dist_norm.rvs(size=100) # 정규확률분포를 따르는 100개 난수생성
plt.hist(sample_norm, bins=50, rwidth=0.5, color='r')
plt.title("samples from normal distribution")
```
